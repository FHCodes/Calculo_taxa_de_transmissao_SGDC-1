# <font size="6" face="Calibri"> 1) Calculo do fator de m√©rito G/T em dB/K do terminal VSAT.

# <font size="5" face="Calibri">Dados usados:

$f$ = frequ√™ncia de transmiss√£o no downlink = 19 GHz 

$d$ = di√¢metro da antena parab√≥lica = 1 metro

$\eta_ùê¥$ =  par√¢metro de efici√™ncia de abertura = 0.8

**<font size="5" face="Calibri">Utilizando a equa√ß√£o:**

$G(dB) = 10log(109,66.f^2.d^2.\eta_A)$

$G(dB) = 10log(109,66.19^2.1^2.0.8)$

obtemos que $G(dB) =$ **45.00645430445064 dB**

Agora podemos calcular o fator de m√©rito utilizando a seguinte express√£o:

$M = G(dB) - 10log(T)$

$M =$ 45.00645430445064 - 10log(160)

$M =$ **22.965254477891396 dB/K**

# <font size="5" face="Calibri">Vari√°veis utilizadas

```python
d = 1       # Diametro da antena parab√≥lica em metros
Œ∑A = 0.8    # Dendo Œ∑A o par√¢metro de efici√™ncia de abertura
f = 19      # Frequ√™ncia de transmiss√£o em GHz
T = 160     # Temperatura de ru√≠do: 160 Kelvin
```

Calculando o ganho da antena em dB

```python
G = 10*log10(109.66*(f**2)*(d**2)*Œ∑A)
```

```python
print(f"Temos que o ganho da antena √© de {G} dB")
```

Calculando o fator de m√©rito do terminal VSAT em dB/K

```python
GT = G - 10*log10(T)
```

```python
print(f"Temos que o fator de m√©rito √© igual √† {GT} dB/K")
```

$r_e$ =  Raio equatorial = 6378.14 Km

Para determinar a distancia da esta√ß√£o terrena ao sat√©lite requer primeiramente o c√°lculo da dist√¢ncia da esta√ß√£o ao centro da Terra **R** que √© encontrado usando a seguinte express√£o:

$R$ = $\sqrt{L^2 + Z^2}$

Sendo $L$ e $Z$ auxiliares para o c√°lculo, sendo calculados das seguintes formas:

$L$ = $  \left(\large\frac{r_e}{\sqrt{1 - e_e^2.\sin^2(L_E)}} + H\right) \cos(L_E)$ 

$L$ = $  \left(\large\frac{6378.14}{\sqrt{1 - 0.08182^2\sin^2(-5.80993)}} +  0.043\right) \cos(-5.80993)$ 

$L = $**6345.6370788682825 Km**

$Z$ = $  \left(\large\frac{r_e(1 - e_e^2)}{\sqrt{1 - e_e^2\sin^2(L_E)}} + H\right) \sin(L_E)$ 

$Z$ = $  \left(\frac{6378.14(1 - 0.08182^2)}{\sqrt{1 - 0.08182^2\sin^2(-5.80993)}} +  0.043\right) \sin(-5.80993)$

$Z = $**-641.3549792057134 Km**

Calculando um angulo auxiliar $\psi_E$ 

$\psi_E$ = $\tan^{-1}(\frac{Z}{L})$

$\psi_E$ = $\tan^{-1}(\frac{-641.3549792057134}{6345.6370788682825})$

$\psi_E = $**-5.771299041466787¬∞**

 $R$ = $\sqrt{L^2 + Z^2}$

$R$ = $\sqrt{(6345.6370788682825)^2 + (-641.3549792057134)^2}$

$R = $**6377.965674575236 km**

Calculando a  $B$, onde $B$ √© a longitude diferencial:

$B = I_e - I_s$

$B = $13.700920000000012¬∞

Finalmente podemos achar a dist√¢ncia da antena ao sat√©lite que √© dada pela f√≥rmula:

$d = \sqrt{R^2 + r_s^2 - 2Rr_s \cos(\psi)cos(B)}$

$d = \sqrt{(6377.965674575236)^2 + (42164.17)^2 - (2)(6377.965674575236)(42164.17) \cos(-5.771299041466787¬∞)cos(13.700920000000012¬∞)}$

$d = $**36036.16548174784 Km**

Agora vamos calcular o angulo de eleva√ß√£o ao sat√©lite, que √© dado pela express√£o:

$\theta = \cos^{-1}\left(\frac{\Large r_e + h_{gso}}{\large d}\sqrt{1 - \cos^2(B)\cos^2(L_e)}\right)$

$\theta = \cos^{-1}\left(\frac{\large 6378.14 + 35786}{\large 36036.16548174784 }\sqrt{1 - \cos^2(13.700920000000012¬∞)\cos^2(-5.80993¬∞)}\right)$

Por fim ser√° calculado o √¢ngulo de azimute ao sat√©lite, mas antes precisamos achar o angulo intermediario $A_i$ para nos auxiliar:

 $A_i = \sin^{-1}\left(\frac{\large\sin(|B|)}{\large\sin(\beta)}\right)  $

onde $\beta$ pode ser calculado pela rela√ß√£o:

$\beta = \cos^{-1}[\cos(B)\cos(L_e)] $

$\beta = \cos^{-1}[\cos(13.700920000000012¬∞)\cos(-5.80993¬∞)] $

Achando $A_i$:

# <font size="5" face="Calibri"> Vari√°veis utilizadas

```python
re = 6378.14    # Raio equatorial medido em KM
hgso = 35786    # Altitude geoestacionaria medido em KM
rs = 42164.17   # Raio geoestacion√°rio medido em KM
ee = 0.08182    # Excentricidade da Terra

Ie = radians(-61.29908)    # Longitude da esta√ß√£o terrena, em graus
Is = radians(-75)           # Longitude do satelite, em graus
Le = radians(-5.80993)     # Latitude da esta√ß√£o terrena, em graus
Ls = 0                     # Latitude do sat√©lite, em graus 
H = 0.043                  # Altitude da esta√ß√£o terrena acima do n√≠vel do mar medido em KM
```

# <font size="5" face="Calibri"> Achando a dist√¢ncia da antena ate o sat√©lite

Achando o valor de **L**

```python
#Simplificando a visualiza√ß√£o do codigo
baixo = sqrt(1 - (ee**2)*sin(Le)**2)
```

```python
L = ((re/baixo)+H)*cos(Le)
```

```python
print(f"Temos que o valor de L √© {L}")
```

Achando o valor de **Z**

```python
#Simplificando a visualiza√ß√£o do codigo
cimaZ = (re*(1-ee**2))
```

```python
Z = ((cimaZ/baixo)+H)*sin(Le)
```

```python
print(f"Temos que o valor de Z √© {Z}")
```

Com **L** e **Z** calculados podemos achar agora o valor de Œ®e

```python
Œ®e = atan(Z/L) # Aqui o valor esta em radianos para efeito de calculos
```

```python
print(f"O valor de Œ®e √© {degrees(Œ®e)}¬∞") # aqui o valor esta em graus para melhor visualiza√ß√£o
```

Calculando **B**, onde **B** √© a longitude diferencial

```python
B = Ie - Is  # Aqui o valor esta em radianos para efeito de calculos
```

```python
print(f"O valor de B √© {degrees(B)}¬∞") # Aqui o valor esta em graus para melhor visualiza√ß√£o
```

Achando **R** que √© a dist√¢ncia da esta√ß√£o ao centro da Terra

```python
R = sqrt(L**2 + Z**2)
```

```python
print(f"O valor encontrado para R foi {R}")
```

**Por fim ser√° calculado a dist√¢ncia da antena at√© o sat√©lite**

```python
d = sqrt(R**2 +rs**2 - 2*R*rs*cos(Œ®e)*cos(B))
```

```python
print(f"A distancia da antena ate o sat√©lite √© de {d} Km")
```

### <font size="5" face="Calibri"> Calculando o √¢ngulo de eleva√ß√£o ao sat√©lite

```python
Œò = acos(((re + hgso)/d)*sqrt(1 - (cos(B)**2)*(cos(Le)**2)))
```

```python
print(f"O valor de Œò √© {degrees(Œò)}¬∞ ") 
```

- **Caso1:** sem chuva: 3 dB

**Caso2:** $ L_0(db) = 11 dB$

$(C/N)_{db} = 49 + 22.965254477891396 - (210.51390811225644 + 11 + 228,6 - 36000000$

Efici√™ncia espectral √© a raz√£o entre a taxa de transmiss√£o do enlace e banda ocupada pelo sinal.

$Ef = \frac{\large R_b}{\large B} $  $\Longleftrightarrow$ $R_b = B.Ef$

Taxa de transmiss√£o **sem chuva**

$R_b = B.Ef_{sc}$

$R_b = (36000000)(3,4)$

$R_b =$**122.40 Mbps**

Taxa de transmiss√£o **chuva mediana**

$R_b = B.Ef_{cm}$

$R_b = (36000000)(2,2)$

$R_b =$**79.20 Mbps**

Taxa de transmiss√£o **chuva forte**

$R_b = B.Ef_{cf}$

$R_b = (36000000)(1,5)$

$R_b =$**54.00 Mbps**

# <font size="5" face="Calibri"> Variaveis utilizadas

```python
Efsc = 3.4      # Efici√™ncia espectral sem chuva
Efcm = 2.2      # Efici√™ncia especytral com chuva mediana
Efcf = 1.5      # Efici√™ncia espectral com chuva forte
B = 36000000    # Banda do transponder em Hz
```

Calculando Rb(Taxa de transmiss√£o)

Temos que a efici√™ncia espectral √© igual a banda multiplicada pela taxa de transmiss√£o, logo podemos calcular Rb como, Rb = B*Ef

```python
Rb = B*Efsc
```

```python
Rb = B*Efcm
```

```python
print(f"m√°xima taxa de transmiss√£o √© {(Rb/1000000):.2f} Mbps")
```

**Chuva forte**

```python
Rb = B*Efcf
```
